```python
import nltk
from stop_words import get_stop_words
from nltk.stem.snowball import SnowballStemmer
from datetime import datetime
import re
import distance
```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
def number_words_query(query):
 try: 
  querywords = nltk.word_tokenize(str(query).lower()) 
  stop_words = get_stop_words('en')
  not_stop_words=[]
  for word in querywords:
   if(word not in stop_words):
    not_stop_words.append(word)
    querylength = len(not_stop_words)
 except:
   querylength=1
 return querylength
```

```python
def number_words_title(query,title,intitle):
 try:
  #print(query,title,intitle, sep="---")
  stemmer = SnowballStemmer("english", ignore_stopwords=True)
  #print("stemmer",stemmer)
  wordsintitle=0
  titlewords = nltk.word_tokenize(str(title).lower())
  #print("titlewords",titlewords)
  stemtitlewords = [stemmer.stem(word) for word in titlewords]
  #print("stemtitlewords", stemtitlewords)
  querywords = nltk.word_tokenize(str(query).lower()) 
  #print("querywords",querywords)
  stemquerywords = [stemmer.stem(word) for word in querywords]
  #print("stemquerywords",stemquerywords)
  stop_words = get_stop_words('en')
  #print("stop_words",stop_words)
  titlewordsnospaces=title.replace(' ','').lower()
  #print("titlewordsnospaces",titlewordsnospaces)
  titlewordsnosigns = re.sub(r'\W+', '', titlewordsnospaces)
  #print("titlewordsnosigns",titlewordsnosigns)
  distanceword=0 
  for word in stemquerywords:
   if(word not in stop_words):
    #print("First IF ",word)
    if(word in titlewordsnospaces or word in titlewordsnosigns):
     #print("in the first if loop for word ",word)
     wordsintitle=wordsintitle+1 
    else:
         for word2 in stemtitlewords:
          distanceword = distance.levenshtein(str(word),str(word2))
          #print("words are word: ",word,"word2: ",word2)    
          #print("distance is ",distanceword)
          procentdistance = distanceword/(len(word)+len(word2)/2)
          #print("procentdistance is ",procentdistance)
          if distanceword<1 or procentdistance<0.15:
           #print("in the second if loop for word2 ",word2)
           wordsintitle=wordsintitle+1
 except: 
  wordsintitle=1
 #print("wordsintitle",wordsintitle)
 return wordsintitle
```

```python
test['numberwordsinUrl'] = test.apply(lambda x: number_words_query(x['url']),axis=1)
test['numberwordsinTitle'] = test.apply(lambda x: number_words_query(x['Title']),axis=1)
test['numberwordsinDesc'] = test.apply(lambda x: number_words_query(x['description']),axis=1)
test['numberwordsinH1'] = test.apply(lambda x: number_words_query(x['h1_contents']),axis=1)
test['numberwordsinquery'] = test.apply(lambda x: number_words_query(x['query']),axis=1)
test['numberwordsofcommonwords_url'] = test.apply(lambda x: number_words_title(x['query'],x['url'],True),axis=1) 
test['numberwordsofcommonwords_title'] = test.apply(lambda x: number_words_title(x['query'],x['url'],True),axis=1) 
test['numberwordsofcommonwords_desc'] = test.apply(lambda x: number_words_title(x['query'],x['url'],True),axis=1) 
test['numberwordsofcommonwords_h1'] = test.apply(lambda x: number_words_title(x['query'],x['url'],True),axis=1) 
test['Percentofcommonwords_url'] = test.apply(lambda x: (x['numberwordsofcommonwords_url']/x['numberwordsinquery'])*100,axis=1)
test['Percentofcommonwords_title'] = test.apply(lambda x: (x['numberwordsofcommonwords_title']/x['numberwordsinquery'])*100,axis=1)
test['Percentofcommonwords_desc'] = test.apply(lambda x: (x['numberwordsofcommonwords_desc']/x['numberwordsinquery'])*100,axis=1)
test['Percentofcommonwords_h1'] = test.apply(lambda x: (x['numberwordsofcommonwords_h1']/x['numberwordsinquery'])*100,axis=1)
```

```python
test['Percentofcommonwords_url'] = test['Percentofcommonwords_url'].round(decimals=3)
test['Percentofcommonwords_title'] = test['Percentofcommonwords_title'].round(decimals=3)
test['Percentofcommonwords_desc'] = test['Percentofcommonwords_desc'].round(decimals=3)
test['Percentofcommonwords_h1'] = test['Percentofcommonwords_h1'].round(decimals=3)
```
